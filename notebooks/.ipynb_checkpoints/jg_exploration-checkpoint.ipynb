```python
# Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read in data

stock_df = pd.read_csv("/Users/johngalvin/Desktop/GitHub/Project2_Fram_Galvin_Bradford/data/processed/historical_stock_clean.csv")
insider_df = pd.read_csv("/Users/johngalvin/Desktop/GitHub/Project2_Fram_Galvin_Bradford/data/processed/insider_txns_clean.csv")
```

```python
# Set indices as Date

stock_df = stock_df.set_index("Date")
insider_df = insider_df.set_index("Date")
```

```python
# Reverse the row order of the insider_df so that earliest date is first

insider_df = insider_df.loc[::-1]
```

# Insider Transactions Exploration

```python
insider_df.head()
```

```python
insider_df.describe()
```

```python
# Determine how many companies had insider_txns in the time period of interest

len(insider_df["Symbol"].unique())
```

```python
# Plot total transactions over time

insider_df.groupby([insider_df.index])["Insider Trading Shares"].count().plot(kind="line", colormap="Set2", figsize=(12,5), legend=False, title="Total Number of Transactions", ylabel="Number of Transactions")
pass
```

```python
# Plot total shares traded per day

insider_df.groupby([insider_df.index])["Insider Trading Shares"].sum().plot(kind="line", figsize=(12,5), legend=False, title="Total Number of Shares Traded", ylabel="Number of Shares")
pass
```

```python
# Top 10 Symbols with the most transactions
# Disregard "LIMIT" entries

insider_df["Symbol"].value_counts()[1:11].plot(kind="bar", colormap="Set2", figsize=(10,5), title="Top 10 Companies, by Transactions", xlabel="Symbol", ylabel="Number of Transactions")
pass
```

```python
# Top 10 symbols with the largest amount of shares traded

insider_df.groupby("Symbol")["Insider Trading Shares"].sum().sort_values(ascending=False)[1:11].plot(kind="bar", figsize=(10,5), title="Top 10 Companies, by Shares", ylabel="Number of Shares")
pass
```

```python
# Determine number of buys ("P") and sells ("S")

insider_df["Buy/Sell"].value_counts().plot(kind="bar", figsize=(7,5), title="Number of Sells (S) and Purchases (P)", ylabel="Number of Transactions", rot=0)
pass
```

```python
# Plot buys ("P") and sells ("S") over time

ax = pd.DataFrame(insider_df.groupby(insider_df.index)["Buy/Sell"].value_counts()).unstack(level=1).plot(figsize=(12,5))
ax.set_title("Sells (S) and Purcahses (P) Over Time")
ax.set_xlabel("Date", labelpad=10)
ax.set_ylabel("Number of Transactions", labelpad=10)
my_labels = ["Purchases", "Sells"]
ax.legend(labels=my_labels)
pass
```

```python
# See which variables are correlated with one another

corr = insider_df.corr("spearman")
matrix = np.triu(corr)
fig, ax = plt.subplots(figsize=(8,7))
sns.heatmap(corr, xticklabels = corr.columns, yticklabels = corr.columns, annot=True, mask=matrix, cmap = sns.diverging_palette(220,20, as_cmap=True), ax=ax)
pass
```

# Combined - Insider / Stock Price Exploration

### Plot 1 - Top 10 Companies, by Shares

```python
# Build a list of the Top 10 Companies, by Shares

top_by_shares = list(insider_df.groupby("Symbol")["Insider Trading Shares"].sum().sort_values(ascending=False)[1:11].index)
```

```python
# Find which day the peak in trading (by shares) occurred on

peak_day = insider_df.groupby([insider_df.index])["Insider Trading Shares"].sum().idxmax()
```

```python
# Plot historical stock prices for the top 10 companies, by shares

ax = stock_df.loc[:, top_by_shares].plot(figsize=(12,7), fontsize=10)
ax.legend(bbox_to_anchor=(1, 0.66), ncol=1)
ax.set_ylabel("Price per Share", fontdict={"fontsize":14}, labelpad=10)
ax.set_xlabel("Date", fontdict={"fontsize":14}, labelpad=10)
ax.set_title("Historical Stock Prices for Top 10 Companies, by Shares", fontdict={"fontsize":14})
ax.axvline(list(stock_df.index).index(peak_day), alpha=0.6, linestyle="--")
plt.annotate("Peak in number of shares traded", (50,3750))
plt.arrow(47, 3750, -15, 0, head_width=40, head_length=3, linewidth=2, color="g")
plt.tight_layout()
pass
```

### Plot 2 - Top 10 Companies, by Transactions

```python
# Build a list of the Top 10 Companies, by Transaction

top_by_txns = insider_df["Symbol"].value_counts()[1:11]
```

```python
# Plot historical stock prices for the top 10 companies, by number of total transactions

ax = stock_df.loc[:, top_by_txns.index].plot(figsize=(12,7), fontsize=10)
ax.legend(bbox_to_anchor=(1, 0.66), ncol=1)
ax.set_ylabel("Number of Transactions", fontdict={"fontsize":14}, labelpad=10)
ax.set_xlabel("Date", fontdict={"fontsize":14}, labelpad=10)
ax.set_title("Historical Stock Prices for Top 10 Companies, by Transactions", fontdict={"fontsize":14})
ax.axvline(list(stock_df.index).index(peak_day), alpha=0.6, linestyle="--")
plt.annotate("Peak in number of shares traded", (40,168))
plt.arrow(39, 168, -5, 0, head_width=2, head_length=3, linewidth=2, color="g")
plt.tight_layout()
pass
```
